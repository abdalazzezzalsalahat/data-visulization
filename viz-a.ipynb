```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('planets')
df.head(5)
```

```python
sns.relplot(x='mass', y='year', hue=df.year.tolist(), data=df);
```

```python
year_mean = df.groupby('orbital_period', as_index=False).mean()

year_mean.head(5)
```

```python
graph = sns.barplot(x="number", y="distance", data=year_mean, ci=None)
```

```python
sns.lmplot(x="mass", y="distance", data=year_mean);
```

```python
graph = sns.relplot(x="mass", y="orbital_period", hue=df.year.tolist(), data=df)
```
