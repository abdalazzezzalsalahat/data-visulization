```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import pyplot
```

```python
df = sns.load_dataset('car_crashes')
df.head(5)
```

```python
# fig, ax = pyplot.subplots(figsize=(60,10))
sns.relplot( x='total', y='speeding', hue=df.speeding.tolist(), data=df)
```

```python
total_df = df.groupby('total', as_index=False).sum()
total_df.head(5)
```

```python
graph = sns.barplot(x="speeding", y="no_previous", data=total_df, ci=None)
```

```python
att_df = df.groupby('total', as_index=False).sum()
att_df.head(5)
```

```python
graph = sns.barplot(x="not_distracted", y="speeding", data=att_df, ci=None)
```

```python
sns.lmplot(x="speeding", y="total", data=att_df);
```
